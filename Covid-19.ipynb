# Time series analyis using ARIMA model

## Prediction for Confirmed Cases
- Separating ConfirmedCases column

```python
con=df.iloc[:,1:2]  ### ConfirmedCases as con
```

```python
con
```

## Visualize the time series

```python
plt.figure(figsize=(10, 6))
plt.plot(con['CasesConfirmed'])
plt.title('COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.show()
```

## Stationarity check
### Perform Augmented Dickey-Fuller test

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(con['CasesConfirmed'], autolag='AIC')

print('p-value:', result[1])

# p-value should be less than or equal to 0.05
if result[1] <= 0.05:
    print("Conclusion: Reject the null hypothesis (Time series is stationary)")
else:
    print("Conclusion: Fail to reject the null hypothesis (Time series is not stationary)")
```

## Differencing for stationarity

```python
con_diff = con['CasesConfirmed'].diff().dropna()

result = adfuller(con_diff)

print('p-value:', result[1])

if result[1] <= 0.05:
    print("Conclusion: Reject the null hypothesis (Time series is stationary)")
else:
    print("Conclusion: Fail to reject the null hypothesis (Time series is not stationary)")
```

### p-value less than 0.05 signifies a stationary timeseries

## Using TimeSeriesSplit to maintain the temporal integrity of the data

```python
from sklearn.model_selection import TimeSeriesSplit

tscv= TimeSeriesSplit(n_splits=5) # for model efficiency

for train_index, test_index in tscv.split(con_diff):
    train_con, test_con = con_diff.iloc[train_index], con_diff.iloc[test_index]
```

```python
train_con, test_con
```

##  ACF and PACF plots for the differenced series


```python
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
plot_acf(train_con)
plt.show()

plot_pacf(train_con)
plt.show()
```

## Importing ARIMA model

#### Assuming values for p and q

```python
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.arima.model import ARIMA

p, d, q = 1, 0, 19

# Train ARIMA model
model = ARIMA(train_con, order=(p, d, q))
results = model.fit()

# Summary of the model
print(results.summary())
```

##  Plotting residuals  for nature of distribution

```python
residuals = results.resid
residuals.plot(kind='kde')
plt.show()
```

## Visualizing model performance based on identified (AR)p and (MA)q values from ACF and PACF plots

```python
forecast_steps = 100 

# Forecast future values
forecast = results.get_forecast(steps=forecast_steps)

# Extract forecasted values and confidence intervals
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Plot the original time series and the forecast
plt.figure(figsize=(15,5))
plt.plot(train_con, label='Observed')
plt.plot(forecast_values, color='red', label='Forecast')
plt.fill_between(confidence_intervals.index, confidence_intervals.iloc[:, 0], confidence_intervals.iloc[:, 1], color='red', alpha=0.3, label='Confidence Intervals')
plt.legend()
plt.show()
```

```python
# Forecast future values on the testing set
forecast = results.get_forecast(steps=len(test_con))
forecast_values = forecast.predicted_mean


# Plot the original time series and the forecast
plt.figure(figsize=(10,5))
plt.plot(train_con, label='Original Time Series')
plt.plot(test_con.index, forecast_values, color='red', label='Forecast')
plt.legend()
plt.show()
```

## Model evaluation

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_con, forecast_values)
print('Mean Absolute Error:', mae)
```

- MAE appears to be large due to large values of the dataset

## Visualization of actual values with forcasted values

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(train_con, label='Train')
plt.plot(test_con, label='Actual Test')
plt.plot(test_con.index, forecast_values, color='red', label='Forecasted Test')
plt.fill_between(confidence_intervals.index, confidence_intervals.iloc[:, 0], confidence_intervals.iloc[:, 1], color='cyan', alpha=0.1, label='Confidence Intervals')
plt.title('ARIMA Model Forecasting')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```
