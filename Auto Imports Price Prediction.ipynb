# _DataPreprocessing_

### _Changing Numerical columns into Numerical datatype_

```python
## Using mappping
mapping = {'two': 2, 'four': 4}
df['Num_of_Doors'] = df['Num_of_Doors'].replace(mapping)
```

```python
cylinders = {'two': 2, 'three': 3, 'four': 4, 'five': 5, 'six': 6, 'eight': 8, 'twelve': 12}
df['Num_of_Cylinders'] = df['Num_of_Cylinders'].replace(cylinders)
```

```python
df.info()
```

### _Handling missing Values_

```python
df.isnull().sum()
```

### _Seperating Numerical and Categorical columns_

##### Numerical Columns

```python
d=df.iloc[:,[0,1,5,9,10,11,12,13,15,16,18,19,20,21,22,23,24,25]]
d
```

### _Box plot_

```python
plt.figure(figsize=(15,30))
plotnumber=1 #counter

for column in d: 
    if plotnumber<=18 :
        ax = plt.subplot(6,3,plotnumber)
        sns.histplot(x=d[column],kde=True)
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
```

### _Heatmap for Correlation_

```python
plt.figure(figsize=(15, 15))
sns.heatmap(d.corr(), annot=True, fmt ='.0%',linewidths=.5, cmap= 'YlGnBu')
plt.show()
```

### _Dropping columns: Normalized Losses, Compression Ratio and Stroke due to redundency_

```python
d.drop(columns=['Compression_Ratio', 'Normalized_Losses', 'Stroke'], inplace=True) ## Dropping due to high Outliers
d.info()
```

### _Replacing Null values with median since data is skewed_

```python
Door_median=d.Num_of_Doors.median()
d.Num_of_Doors.fillna(Door_median, inplace=True)
Door_median
```

```python
B_median=d.Bore.median()
d.Bore.fillna(B_median, inplace=True)
B_median
```

```python
H_median=d.Horse_Power.median()
d.Horse_Power.fillna(H_median, inplace=True)
H_median
```

```python
RPM_median= d.Peak_RPM.median()
d.Peak_RPM.fillna(RPM_median, inplace=True)
RPM_median
```

```python
d.isnull().sum()
```

## _Outlier Handling_

### _Box plot_

```python
plt.figure(figsize=(15,30))
plotnumber=1 

for column in d: 
    if plotnumber<=18 :
        ax = plt.subplot(6,3,plotnumber)
        sns.boxplot(x=d[column],hue=d['Price'])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
```

### _Replacing outlier values with median keeping upper bound at approximately 85% to avoid underfitting_

```python
d['Price']=np.where(d['Price']>=40000,d['Price'].median(),d['Price'])
```

```python
sns.boxplot(x=d.Price)
```

## _Seperating Categorical columns_

```python
d_cat_label=df.iloc[:,[2,3,4,6,7,8,14,17]]     ## For Label Encoding
d_cat_label
```

### _Using Label encoding for Linear Model and One Hot encoding for other models on Categorical Data_

#### Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
d_cat_label.Make = label_encoder.fit_transform(d_cat_label.Make)
d_cat_label.Fuel_Type = label_encoder.fit_transform(d_cat_label.Fuel_Type)
d_cat_label.Aspiration = label_encoder.fit_transform(d_cat_label.Aspiration)
d_cat_label.Body_Style = label_encoder.fit_transform(d_cat_label.Body_Style)
d_cat_label.Drive_Wheels = label_encoder.fit_transform(d_cat_label.Drive_Wheels)
d_cat_label.Engine_Location = label_encoder.fit_transform(d_cat_label.Engine_Location)
d_cat_label.Engine_Type = label_encoder.fit_transform(d_cat_label.Engine_Type)
d_cat_label.Fuel_System = label_encoder.fit_transform(d_cat_label.Fuel_System)
d_cat_label
```

#### OneHot Encoding

```python
d_cat_one=df.iloc[:,[2,3,4,6,7,8,14,17]]
d_cat_one = pd.get_dummies(d_cat_one, columns=['Make', 'Fuel_Type', 'Aspiration', 'Body_Style', 'Drive_Wheels', 'Engine_Location', 'Engine_Type', 'Fuel_System'], drop_first=True)
d_cat_one
```

### _Using MinMaxScaler for Scaling Data_

```python
from sklearn.preprocessing import MinMaxScaler
scale= MinMaxScaler()
d_num_scaled=scale.fit_transform(d)
d_num_scaled = pd.DataFrame(d_num_scaled, columns=d.columns)
d_num_scaled
```

### _Making a New DataFrame storing the Scaled Data & Label encoded data for Linear Model_

```python
df_linear= pd.concat([d_cat_label,d_num_scaled],axis=1)
df_linear.info()
```

### _Making a seperate DataFrame for Non-Linear Models_

```python
dff= pd.concat([d_cat_one,d_num_scaled],axis=1)
dff.info()
```

# _Model Creation_

## _Linear Regression_

### _Target and Feature seperation_

```python
from sklearn.model_selection import train_test_split
```

```python
Xl=df_linear.iloc[:,:-1]
yl=df_linear.Price
```

```python
Xl_train,Xl_test,yl_train,yl_test=train_test_split(Xl,yl,test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lmodel=LinearRegression()
lmodel.fit(Xl_train, yl_train)
```

```python
yl_pred=lmodel.predict(Xl_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print(mean_absolute_error(yl_test,yl_pred))
print(r2_score(yl_test,yl_pred))
```

## _Random Forest_

```python
X=dff.iloc[:,:-1]
y=dff.Price
Xr=X
yr=y
```

```python
Xr_train,Xr_test,yr_train,yr_test=train_test_split(Xr,yr,test_size=0.2, random_state=50)
```

```python
#model creation 
from sklearn.ensemble import RandomForestRegressor
model_r = RandomForestRegressor(random_state=39)
model_r.fit(Xr_train, yr_train)
```

```python
yr_pred=model_r.predict(Xr_test)
```

```python
print(mean_absolute_error(yr_test,yr_pred))
print(r2_score(yr_test,yr_pred))
```

## _HyperTuning the Model for better Accuracy using GridSearchCV for finding BestParams_

## _RandomForest using BestParams_

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20],  
    'min_samples_split': [2, 5, 10], 
    'min_samples_leaf': [1, 4, 5]      
}

# Perform Grid Search with 5-fold cross-validation
gscv = GridSearchCV(estimator=model_r, param_grid=param_grid, cv=5, verbose=2)
gscv.fit(Xr,yr)
gscv.best_params_
```

####  _Best Params after using GridSearchCV_
- 'max_depth': None,
- 'min_samples_leaf': 4,
- 'min_samples_split': 10,
- 'n_estimators': 100
